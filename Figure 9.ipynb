```python
import numpy as np
import scipy.io as sio
from scipy import interpolate
from scipy import signal
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib.colors import LinearSegmentedColormap
%matplotlib inline
```

```python
def create_cmap(color1, color2, color3):
    colors = [color1, color2, color3]
    cmap_name = 'list'
    n_bins = 100
    cmap = LinearSegmentedColormap.from_list(cmap_name, colors, N=n_bins)
    return cmap

def build_hankel(data, rows, cols):
    X = np.empty((rows, cols))
    for k in range(rows):
        X[k, :] = data[k:cols + k]
    return X 

def HAVOK(X, dt, r, norm, center=False, return_uv=False):
    if (center): 
        m = X.shape[0]
        X̄ = X - X[m//2,:]
        U, Σ, Vh = np.linalg.svd(X̄,full_matrices=False)
    else:
        U, Σ, Vh = np.linalg.svd(X,full_matrices=False)
    V = Vh.T
    polys = true_polys(X.shape[0], dt, r, center)
    for _i in range(r):
        if (np.dot(U[:,_i], polys[:,_i]) < 0):
            U[:,_i] *= -1
            V[:,_i] *= -1
    V1 = V[:-1,:r]
    V2 = V[1:,:r]
    A = (V2.T @ np.linalg.pinv(V1.T) - np.eye(r)) / (norm * dt)
    if (return_uv): 
        return A, U, V
    return A

def sHAVOK(X, dt, r, norm):
    X1 = X[:,:-1]
    X2 = X[:,1:]
    U1,_,Vh1 = np.linalg.svd(X1,full_matrices=False)
    U2,_,Vh2 = np.linalg.svd(X2,full_matrices=False)
    V1 = Vh1.T
    V2 = Vh2.T
    polys = true_polys(X.shape[0], dt, r, center=False)
    for _i in range(r):
        if (np.dot(U1[:,_i], polys[:,_i]) < 0):
            V1[:,_i] *= -1
        if (np.dot(U2[:,_i], polys[:,_i]) < 0):
            V2[:,_i] *= -1
    A = ((V2.T @ V1)[:r,:r] - np.eye(r)) / (norm * dt)
    return A

def true_polys(rows, dt, r, center): 
    m = rows // 2
    Ut = np.linspace(-m*dt, m*dt, rows)
    poly_stack = []
    for j in range(r):
        if (center):
            poly_stack.append(Ut ** (j + 1))
        else: 
            poly_stack.append(Ut ** j)
    poly_stack = np.vstack(poly_stack).T
    Q = np.empty((rows, r)) # Perform Gram-Schmidt
    for j in range(r): 
        v = poly_stack[:, j]
        for k in range(j - 1): 
            r_jk = Q[:, k].T @ poly_stack[:, j]
            v -= (r_jk * Q[:, k])
        r_jj = np.linalg.norm(v)
        Q[:, j] = v / r_jj
    return Q

def reconstruct_v(A, Vh_full, r, dt):
    Ā = A[:-1, :-1]
    B = A[:-1, -1].reshape(-1, 1)
    t0 = np.arange(Vh_full.shape[1]) * dt 
    sys = signal.StateSpace(Ā, B, np.eye(r-1), np.zeros(r-1).reshape(-1, 1))
    tout, y, x = signal.lsim(sys, Vh_full[r-1, :], t0, X0=Vh_full[:r-1, 0])
    Vh_rec = y.T
    return Vh_rec

# Define colormap 
cmap = create_cmap('tab:blue', 'white', 'tab:orange')
```

# Double Pendulum

```python
# Get double pendulum data 
θ2 = sio.loadmat('Stady_Data.mat_Final_Result_17_9_5_4.996.mat')['The2'][::10]
pendulum_dt = 0.001
xdata = np.sin(θ2).flatten()[1500:]

# Define parameters 
pendulum_r = [5, 5, 4, 4]
rows = 101

pendulum_short = dict()
pendulum_full = dict()
pendulum_recon = dict()
pendulum_reconf = dict()
pendulum_short['n'] = 700
pendulum_full['n'] = 8500
pendulum_recon['n'] = 650
pendulum_reconf['n'] = 8500

for i, elem in enumerate((pendulum_short, pendulum_full, pendulum_recon, pendulum_reconf)):
    elem['xdata'] = xdata[:elem['n']]
    elem['H'] = build_hankel(elem['xdata'],rows,elem['xdata'].size-rows+1)
    elem['Vh'] = np.linalg.svd(elem['H'], full_matrices=False)[2]
    elem['A1'], U, V = HAVOK(elem['H'],pendulum_dt,pendulum_r[i],1,return_uv=True)
    elem['Vh'] = V.T
    elem['A2'] = sHAVOK(elem['H'],pendulum_dt,pendulum_r[i],1)
    elem['ω1'] = np.linalg.eig(elem['A1'])[0]
    elem['ω2'] = np.linalg.eig(elem['A2'])[0]
    
# Plot results 
plt.figure(figsize=(18,3),dpi=200)

# Picture
plt.subplot(1,5,1)
img = mpimg.imread('doublependulum.png')
plt.imshow(img)
plt.axis('off')
plt.xticks([])
plt.yticks([])

# Time series 
ax = plt.subplot(1,5,2)
y_shift = 0.4
plt.plot(np.arange(pendulum_full['n'])*pendulum_dt, pendulum_full['xdata'], color='gray')
plt.plot(np.arange(pendulum_short['n'])*pendulum_dt, pendulum_short['xdata'],color='black')
plt.ylim(np.min(pendulum_full['xdata']*1.3),np.max(pendulum_full['xdata']*1.3))
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
plt.ylabel(r'$\sin(\theta_2)$', fontsize=14, color='white')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

# HAVOK dynamics matrix 
plt.subplot(1,5,3)
plt.imshow(pendulum_short['A1'],cmap=cmap)
plt.xticks([])
plt.yticks([])
plt.grid('on',color='k')

# sHAVOK dynamics matrix 
plt.subplot(1,5,4)
plt.imshow(pendulum_short['A2'],cmap=cmap)
plt.xticks([])
plt.yticks([])

# Eigenvalues 
ax = plt.subplot(1,5,5)
ax.spines['bottom'].set_position('zero')
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_color('gray')
ax.spines['left'].set_color('gray')
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
plt.xticks([])
plt.yticks([])
plt.plot(pendulum_short['ω1'].real,pendulum_short['ω1'].imag,'o',color='teal',markersize=14,label='HAVOK',alpha=0.5)
plt.plot(pendulum_short['ω2'].real,pendulum_short['ω2'].imag,'o',color='maroon',markersize=14,label='sHAVOK',alpha=0.5)
plt.plot(pendulum_full['ω1'].real,pendulum_full['ω1'].imag,'+',color='k',markersize=14,mew=2,label='true',linewidth=20)
plt.tight_layout()
plt.savefig('pdf_files/fig9a.pdf', bbox_inches='tight')
```

# Measles

```python
# Get measles data 
measles = np.loadtxt('nycmeas.dat')
tspan = measles[:, 0] # time (in years)
cases = measles[:, 1] / 1000 # number of cases (in thousands)

# Interpolate measles data
samples = 20001
tspan_interp = np.linspace(tspan[0], tspan[-1], samples)
tck = interpolate.splrep(tspan, cases)
cases_interp = interpolate.splev(tspan_interp, tck)
measles_dt = tspan_interp[1] - tspan_interp[0]

# Define parameters 
measles_r = [6, 6, 5, 5]
rows = 51

measles_short = dict()
measles_full = dict()
measles_recon = dict()
measles_reconf = dict()
measles_short['n'] = 280
measles_full['n'] = samples
measles_recon['n'] = 280
measles_reconf['n'] = samples

for i, elem in enumerate((measles_short, measles_full, measles_recon, measles_reconf)):
    elem['xdata'] = cases_interp[:elem['n']]
    elem['H'] = build_hankel(elem['xdata'],rows,elem['xdata'].size-rows+1)
    elem['Vh'] = np.linalg.svd(elem['H'], full_matrices=False)[2]
    elem['A1'], U, V = HAVOK(elem['H'],measles_dt,measles_r[i],1,return_uv=True)
    elem['Vh'] = V.T
    elem['A2'] = sHAVOK(elem['H'],measles_dt,measles_r[i],1)
    elem['ω1'] = np.linalg.eig(elem['A1'])[0]
    elem['ω2'] = np.linalg.eig(elem['A2'])[0]
    
# Plot results 
plt.figure(figsize=(16,3),dpi=200)

# Picture
plt.subplot(1,5,1)
img = mpimg.imread('measles.png')
plt.imshow(img)
plt.axis('off')
plt.xticks([])
plt.yticks([])

# Time series 
plt.subplot(1,5,2)
y_shift = 0
plt.plot(tspan_interp[:8200], measles_full['xdata'][:8200] + y_shift,color='gray')
plt.plot(tspan_interp[:len(measles_short['xdata'])], measles_short['xdata'] + y_shift,color='black')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
plt.xticks([1930,1940],fontsize=20)
plt.yticks([0,10,20],fontsize=20)

# HAVOK dynamics matrix 
plt.subplot(1,5,3)
plt.imshow(measles_short['A1'],cmap=cmap,vmin=-20,vmax=20)
plt.xticks([])
plt.yticks([])
plt.grid('on',color='k')

# sHAVOK dynamics matrix 
plt.subplot(1,5,4)
plt.imshow(measles_short['A2'],cmap=cmap,vmin=-20,vmax=20)
plt.xticks([])
plt.yticks([])

# Eigenvalues 
ax = plt.subplot(1,5,5)
ax.spines['bottom'].set_position('zero')
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_color('gray')
ax.spines['left'].set_color('gray')
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
plt.xticks([])
plt.yticks([])
plt.plot(measles_short['ω1'].real,measles_short['ω1'].imag,'o',color='teal',markersize=14,label='HAVOK',alpha=0.5)
plt.plot(measles_short['ω2'].real,measles_short['ω2'].imag,'o',color='maroon',markersize=14,label='sHAVOK',alpha=0.5)
plt.plot(measles_full['ω1'].real,measles_full['ω1'].imag,'+',color='k',markersize=14,mew=2,label='true')

plt.tight_layout()
plt.savefig('pdf_files/fig9b.pdf')
```
